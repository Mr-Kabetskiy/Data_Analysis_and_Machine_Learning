```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('./content/perfomance.txt', index_col=0)
```

#### 2.17.1 Постройте круговую диаграмму для колонки gender.

```python
df.gender.value_counts().plot.pie()
```

#### 2.17.2 Постройте круговую диаграмму для колонки group.

```python
df.group.value_counts().plot.pie();
```

#### 2.17.3 Постройте следующую столбчатую диаграмму:
Два столбца - люди не проходившие подготовку к тесту и люди проходившие ее;
Высота столбцов - средний их балл по математике.

```python
df[['test preparation course', 'math score']].groupby('test preparation course', as_index=False)[
    'math score'].mean().plot.bar(x='test preparation course', y='math score', color=['r', 'g']);
```

```python
t = df[['test preparation course', 'math score']].groupby('test preparation course', as_index=False)[
    'math score'].mean()
sns.barplot(t, x='test preparation course', y='math score');
```

#### 2.17.4 Отобразите любым удобным способом средние оценки по writing для каждой группы

```python
t1 = df.groupby('group', as_index=False)['writing score'].mean()
```

```python
sns.barplot(t1, x='group', y='writing score');
```

```python
sns.barplot(df, x='group', y='writing score', hue='test preparation course', orient='v');
```

#### 2.17.5 Для каждой пары колонок с оценками, отобразите scatterplot - точечную диаграмму

```python
df.columns
```

```python
fig, ax = plt.subplots(1, 3, figsize=(16, 5), sharey=True)
ax[0].scatter(df['math score'], df['reading score'], color='pink')
ax[1].scatter(df['math score'], df['writing score'], color='y')
ax[2].scatter(df['writing score'], df['reading score'], color='green')
plt.show()
```

#### 2.17.6 Отобразите 3d scatterplot для колонок с оценками

```python
import plotly.express as px

px.scatter_3d(df, x='math score', y='reading score', z='writing score');
```
